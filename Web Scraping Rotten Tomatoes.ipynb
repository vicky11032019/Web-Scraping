```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url="https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/2/"

r=requests.get(base_url)

r
```

```python
html=r.content
```

```python
soup=BeautifulSoup(html,"lxml")
```

```python
soup
```

```python
divs=soup.find_all('div',attrs={'class':"col-sm-18 col-full-xs countdown-item-content"})

divs
```

```python
heading= [i.find('h2')  for i in divs]
```

```python
heading[0]
```

```python
 title= [ i.find('a').string   for i in heading]
```

```python
title
```

```python
year= [ i.find('span',attrs={'class':"subtle start-year"}).string   for i in heading]
```

```python
year
```

we need to remove the brackets

```python
 year=  [i.strip("()")   for i in year]
```

```python
year
```

 now converting year into integers

```python
 year=  [int(i)   for i in year]
```

```python
year
```

```python
rating= [ i.find('span',attrs={'class':"tMeterScore"}).string   for i in heading]
```

```python
rating
```

```python
cc= [i.find('div',attrs={'class':"info critics-consensus"})   for i in divs]

#under each film name, there is only one critic consensun, so we are using find and not find_all
```

```python
cc
```

```python
[ i.text  for i in cc]

# with text function, we are pulling only the text data from each divs
```

removing the common text "Critics Consensus: "

```python
common_text='Critics Consensus: '
```

```python
 [[ i.text[len(common_text): ]  for i in cc]]
```

Cast Info:

```python
cast= [i.find('div',attrs={'class':"info cast"})   for i in divs]
```

```python
cast
```

```python
cast[0]

#inspecting the first item, means first movie
```

```python
cast[0].text
```

```python
[i.text for i in cast]
```

```python
common_cast='\nStarring: '

final_cast=[ i.text[len(common_cast): ]  for i in cast]
```

```python
final_cast[3]
```

Pushing the data into DataFrame

```python
import pandas as pd
```

```python
movies=pd.DataFrame()

movies['name']=title
movies['year']=year
movies['rating']=rating
movies['Casts']=final_cast
```
