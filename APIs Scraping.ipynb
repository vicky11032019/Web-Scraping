# iTunes API Scraping

```python
import requests

base_url="https://itunes.apple.com/search?"

r=requests.get(base_url,params={'term':'titanic','country':'us'})

r
```

```python
t=r.json()
```

```python
type(t)
```

```python
t.keys()
```

```python
t['results'][1]
```
